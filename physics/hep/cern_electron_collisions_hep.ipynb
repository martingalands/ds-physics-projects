```python
df['pt_balance_x'] = df.px1 + df.px2
df['pt_balance_y'] = df.py1 + df.py2
```

The transverse momenta are expected to be centered around zero. Significant deviations from zero may indicate the presence of additional particle(s), as momentum conservation requires the vector sum of momentum to vanish in each direction when all particles involved in the interaction are taken into account.

```python
df.pt_balance_x.hist(bins=100)
df.pt_balance_y.hist(bins=100)
plt.title('')
plt.legend()
```

```python
p1 = np.sqrt(df.px1**2 + df.py1**2 + df.pz1**2)
p2 = np.sqrt(df.px2**2 + df.py2**2 + df.pz2**2)

df['mass1_sqrt'] = df.E1**2 - p1**2
df['mass2_sqrt'] = df.E2**2 - p2**2

np.allclose(p1, df.mass1_sqrt)
np.allclose(p2, df.mass2_sqrt)
```

Checking angles (calculated vs dataset) for dataset reconstruction annomalies.

```python
phi1_calc = np.arctan2(df.py1, df.px1)
phi2_calc = np.arctan2(df.py2, df.px2)

np.allclose(df.phi1, phi1_calc)
np.allclose(df.phi2, phi2_calc)
```

What's the proportion between same charge collisions and different charge collisions?

```python
(df.Q1 + df.Q2).value_counts(normalize=True)
```

This makes me think... What about the M histogram for this discrimination of charges?

```python
other_df = df[(df.Q1 + df.Q2) == 0]
same_df = df[(df.Q1 + df.Q2) != 0]

plt.figure(figsize=(10,6))

plt.hist(other_df['M'].dropna(), bins=150, alpha=0.7, density=True, label='Opposite sign collisions')

plt.hist(same_df['M'].dropna(), bins=150, alpha=0.7, density=True, label='Same sign collisions')

plt.xlabel('Invariant mass (M) [GeV]')
plt.ylabel('Density')
plt.grid(alpha=.7)
plt.legend()
plt.show()
```

So, same sign collisions don't show the peaks or bumps found previously wich means those are particles only originated by $e_{+} e_{-}$ collisions.
